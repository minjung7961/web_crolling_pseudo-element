## 크롤링 핵심 코드 패턴으로 이해하기

```python
# 라이브러리 임포트
import requests 
from bs4 import BeautifulSoup


URL = 'https://namu.wiki/w/%EB%A7%88%EC%9D%B8%ED%81%AC%EB%9E%98%ED%94%84%ED%8A%B8/%EC%95%84%EC%9D%B4%ED%85%9C/%EC%96%91%EC%A1%B0'
# 웹페이지 가져오기
res = requests.get(URL)

#웹페이지 파싱하기
soup = BeautifulSoup(res.text, 'html.parser')

#필요한 데이터 추출하기
mydata = soup.find('div','wiki-macro-toc::before')

#추출한 데이터 활용하기
print(mydata)
```

## selenium 으로 크롬브라우저창 열기

```python
from selenium import webdriver
driver = webdriver.Chrome()
url = 'https://namu.wiki/w/%EB%A7%88%EC%9D%B8%ED%81%AC%EB%9E%98%ED%94%84%ED%8A%B8/%EC%95%84%EC%9D%B4%ED%85%9C/%EC%96%91%EC%A1%B0'
driver.get(url)
script = "return window.getComputedStyle(document.querySelector('#toc'),':before').getPropertyValue('content')"
print(driver.execute_script(script))
```
